```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('postgresql://jayz:localhost@localhost:5432/covid1')
```

```python
covid = pd.read_csv('./COVID1/corona.csv')
```

```python
covid.to_sql('covid', engine, index=False)
```

```python
query = 'SELECT * FROM covid;'
df = pd.read_sql(query, engine)

df.head(100)
```

```python
query = 'SELECT gender,count(*) FROM covid group by 1;'
df = pd.read_sql(query, engine)

df
```

```python
query = 'SELECT corona_result,count(*) FROM covid group by 1;'
df = pd.read_sql(query, engine)

df
```
